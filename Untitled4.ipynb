```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
replicates = pd.read_table('SG_molcell/genes.read_group_tracking')
replicates
```

```python
SG_replicates = replicates[replicates['condition'] == 'SG']
total_replicates = replicates[replicates['condition'] == 'TOT']
```

```python
SG_rep1 = SG_replicates[SG_replicates['replicate'] == 1]
SG_rep2 = SG_replicates[SG_replicates['replicate'] == 2]
SG_rep0 = SG_replicates[SG_replicates['replicate'] == 0]

TOT_rep1 = total_replicates[total_replicates['replicate'] == 1]
TOT_rep2 = total_replicates[total_replicates['replicate'] == 2]
TOT_rep0 = total_replicates[total_replicates['replicate'] == 0]

rep1 = SG_rep1['FPKM']
rep2 = SG_rep2['FPKM']
rep0 = SG_rep0['FPKM']
totrep1 = TOT_rep1['FPKM']
totrep2 = TOT_rep2['FPKM']
totrep0 = TOT_rep0['FPKM']

rep1 = rep1.as_matrix()
rep2 = rep2.as_matrix()
rep0 = rep0.as_matrix()
totrep1 = totrep1.as_matrix()
totrep2 = totrep2.as_matrix()
totrep0 = totrep0.as_matrix()

FPKM = pd.DataFrame({'rep1': rep1, 'rep2': rep2, 'rep0': rep0, 'tot1':totrep1, 'tot2' : totrep2, 'tot0':totrep0})
z = np.abs(stats.zscore(FPKM))
FPKM = FPKM[(z < 3).all(axis=1)]
x = FPKM.corr()
x ** 2
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline
x = np.log(FPKM['rep1'])
y = np.log(FPKM['rep2'])
z = np.log(FPKM['rep0'])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
new_plot = ax.scatter(x, y, z, c = 'black', marker = '.')

ax.set_xlabel('replicate 1')
ax.set_ylabel('replicate 2')
ax.set_zlabel('replicate 3')
ax.set_xlim3d(-10, 10, 5)
ax.set_ylim3d(-10, 10, 5)
ax.set_zlim3d(-10,10, 5)
ax.set_xticks([-10, -5, 0, 5, 10])
ax.set_yticks([-10, -5, 0, 5, 10])
ax.set_zticks([-10, -5, 0, 5, 10])
figure = new_plot.get_figure()

figure.savefig('/Users/tyler-matheny/Desktop/3Dstress.png', dpi =400)
```
