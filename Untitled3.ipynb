```python
codons = pd.read_table('METRICS/newCLIP.txt')
RNA_types = pd.read_table('METRICS/type.txt')
SG_metrics = pd.read_csv('sgtranscriptome.csv')
SG_metrics = SG_metrics.rename(index=str, columns={"test_id": "ENSG"})
RNA_types = RNA_types.rename(index=str, columns={"Gene stable ID": "ENSG"})

codons = codons.merge(RNA_types, on = 'ENSG')
codons = codons.loc[codons['Gene type'] == 'protein_coding']
codons = codons.drop_duplicates(subset= ['ENSG'])
codons = codons.merge(SG_metrics, on = 'ENSG', how ='left')
codons = codons.merge(SG_metrics, on = 'ENSG', how ='left')
codons = codons.fillna(0) 
pd.set_option('display.max_columns', 500)
codons
```
