```python
import pandas as pd
import numpy as np
import wget
import os as os
import zipfile

# Download the data
url_1 = 'http://www.kazusa.or.jp/codon/current/species/9606' 
def getdata(url):
    directory = os.getcwd()
    wget.download(url, directory)
getdata(url_1)
```

```python
gene_names = []
codon_list = []
with open('9606') as inputfile:
    for eachLine in inputfile:
        if eachLine.startswith('>'):
            result = eachLine
            result.split('\\')
            result = result.split('\\')
            result = result[0]
            result = result.split('>')
            result = result[1]
            gene_names.append(result)
            codons = next(inputfile)
            codons = codons.split("\n")
            codons = codons[0]
            codon_list.append(codons)
df1 = pd.DataFrame(np.array(gene_names))
df1.columns = ['GenBank Nucleotide Accession']
geneconversion = pd.read_table('/Users/tyler-matheny/Desktop/geneName.txt')
df1 = df1.merge(geneconversion, on = ['GenBank Nucleotide Accession'], how = 'left')
df = pd.DataFrame(np.array(codon_list))
df = df[0].str.split(expand = True)
df2 = pd.concat([df1,df], axis = 1, ignore_index=True)
df2.columns = ["Gene", 'ENSEMBL', 'CGA', 'CGC', 'CGG', 'CGU', 'AGA', 'AGG', 'CUA', 'CUC', 'CUG', 'CUU', 'UUA', 'UUG', 'UCA', 'UCC', 'UCG', 'UCU', 'AGC', 'AGU', 'ACA', 'ACC', 'ACG', 'ACU', 'CCA', 'CCC', 'CCG', 'CCU', 'GCA', 'GCC', 'GCG', 'GCU', 'GGA', 'GGC', 'GGG', 'GGU', 'GUA', 'GUC', 'GUG', 'GUU', 'AAA', 'AAG', 'AAC', 'AAU', 'CAA', 'CAG', 'CAC', 'CAU', 'GAA', 'GAG', 'GAC', 'GAU', 'UAC', 'UAU', 'UGC', 'UGU', 'UUC', 'UUU', 'AUA', 'AUC', 'AUU', 'AUG', 'UGG', 'UAA', 'UAG', 'UGA']
all_codons = df2[['CGA', 'CGC', 'CGG', 'CGU', 'AGA', 'AGG', 'CUA', 'CUC', 'CUG', 'CUU', 'UUA', 'UUG', 'UCA', 'UCC', 'UCG', 'UCU', 'AGC', 'AGU', 'ACA', 'ACC', 'ACG', 'ACU', 'CCA', 'CCC', 'CCG', 'CCU', 'GCA', 'GCC', 'GCG', 'GCU', 'GGA', 'GGC', 'GGG', 'GGU', 'GUA', 'GUC', 'GUG', 'GUU', 'AAA', 'AAG', 'AAC', 'AAU', 'CAA', 'CAG', 'CAC', 'CAU', 'GAA', 'GAG', 'GAC', 'GAU', 'UAC', 'UAU', 'UGC', 'UGU', 'UUC', 'UUU', 'AUA', 'AUC', 'AUU', 'AUG', 'UGG', 'UAA', 'UAG', 'UGA']]
optimal_codons = df2[['CAC', 'UAC', 'UUC', 'UGC', 'GAC', 'AAC', 'AUC', 'GCC', 'ACG', 'GGC', 'CCC', 'AGC', 'AUG', 'UGG','CAG','GAG','AAG','GUG', 'CUG', 'AGG', 'UGA', 'AGA']]
optimal_codons = optimal_codons.apply(pd.to_numeric)
all_codons = all_codons.apply(pd.to_numeric)
optimal_codons['sum'] = optimal_codons.sum(axis = 1)
all_codons['all_sum'] = all_codons.sum(axis = 1)
df_opt = pd.concat([df1,optimal_codons['sum']], axis = 1, ignore_index=True)
df_opt = pd.concat([df_opt,all_codons['all_sum']], axis = 1, ignore_index=True)
df_opt.columns = ["Gene", 'ENSEMBL', 'optimal', 'total']
df_opt['percent_optimal'] = df_opt['optimal']/df_opt['total']
```

```python
df_opt.to_csv('METRICS/codon_optimality.csv')
```
